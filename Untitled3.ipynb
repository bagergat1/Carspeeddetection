```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
inf=pd.read_excel("kayitlarx.xlsx")
inf.pop("Unnamed: 0")
```

```python
inf.drop(0,axis=1,inplace=True)
```

```python
inf.iloc[1]
```

```python
inf
```

```python
plt.subplot(3,1,1)
plt.bar([i for i in range(len(np.array(inf.iloc[1])))],np.array(inf.iloc[1]),color="red")
plt.xlabel("Araç Sayısı")
plt.ylabel("Araçların Hızı")
plt.title("Araçların Hız Grafikleri")

plt.subplot(3,1,2)
plt.bar([i for i in range(len(np.array(inf.iloc[2])))],np.array(inf.iloc[2]),color="green")
plt.xlabel("Araç Sayısı")
plt.ylabel("Sınırı Aşma Oranları (%)")
plt.title("Araçların Hız Sınırını Aşma Oranları")


plt.subplot(3,1,3)
plt.bar([i for i in range(len(np.array(inf.iloc[3])))],np.array(inf.iloc[3]),color="blue")
plt.xlabel("Araç Sayısı")
plt.ylabel("Ceza Oranları (TL)")
plt.title("Araçlara Kesilen Ceza Oranları")
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].bar([i for i in range(len(np.array(inf.iloc[1])))],np.array(inf.iloc[1]),color="red")
plt.xlabel("Araç Sayısı")
plt.ylabel("Araçların Hızı")
plt.title("Araçların Hız Grafikleri")

ax[0,1].bar([i for i in range(len(np.array(inf.iloc[1])))],np.array(inf.iloc[1]),color="red")
plt.xlabel("Araç Sayısı")
plt.ylabel("Araçların Hızı")
plt.title("Araçların Hız Grafikleri")

ax[1,0].bar([i for i in range(len(np.array(inf.iloc[1])))],np.array(inf.iloc[1]),color="red")
plt.xlabel("Araç Sayısı")
plt.ylabel("Araçların Hızı")
plt.title("Araçların Hız Grafikleri")

ax[1,1].bar([i for i in range(len(np.array(inf.iloc[1])))],np.array(inf.iloc[1]),color="red")
plt.xlabel("Araç Sayısı")
plt.ylabel("Araçların Hızı")
plt.title("Araçların Hız Grafikleri")

fig.tight_layout(pad=5.0)
 
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
fig.tight_layout() # Or equivalently,  "plt.tight_layout()"
plt.subplots_adjust(hspace=0.2, wspace=0.2)
axes[0,0].bar([i for i in range(len(np.array(inf.iloc[1])))],np.array(inf.iloc[1]),color="red")
axes[0,0].set_xlabel("Araç ID")
axes[0,0].set_ylabel("Araçların Hızı")
axes[0,0].set_title("Araçların Hız Grafikleri")

axes[0,1].bar([i for i in range(len(np.array(inf.iloc[2])))],np.array(inf.iloc[2]),color="green")
axes[0,1].set_xlabel("Araç ID")
axes[0,1].set_ylabel("Araçların Hızı")
axes[0,1].set_title("Araçların Hız Grafikleri")
fig
axes[1,0].bar([i for i in range(len(np.array(inf.iloc[3])))],np.array(inf.iloc[3]),color="brown")
axes[1,0].set_xlabel("Araç ID")
axes[1,0].set_ylabel("Araçların Hızı")
axes[1,0].set_title("Araçların Hız Grafikleri")

axes[1,1].bar([i for i in range(len(np.array(inf.iloc[4])))],np.array(inf.iloc[4]),color="yellow")
axes[1,1].set_xlabel("Araç ID")
axes[1,1].set_ylabel("Araçların Hızı")
axes[1,1].set_title("Araçların Hız Grafikleri")

fig.show()
```
